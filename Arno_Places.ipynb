```python
import pandas as pd
```

```python
df = pd.read_csv('data/geo/stazioni.csv', ';')
```

```python
df.head()
```

```python
def search_for(nome, unit=None):
    in_name = df['Nome'].str.contains(nome)
    in_comune = df['Comune'].str.contains(nome)
    mydf = df[in_name | in_comune]
    return mydf if unit is None else mydf[mydf['Strumento'] == unit]
```

```python
import pandas as pd
import missingno as msno
import matplotlib.pyplot as plt
```

```python
def import_data(kind, name):
    fn = f"../../data/raw/{kind}/{name}.csv"
    df = pd.read_csv(fn)
    df = df[~df['Date'].isnull()]
    df.index = pd.to_datetime(df['Date'], format='%d/%m/%Y')
    df.index.name = None
    df.index.freq = 'D'
    return df.drop('Date', axis=1)

auser = import_data('aquifer', 'auser')
doganella = import_data('aquifer', 'doganella')
luco = import_data('aquifer', 'luco')
petrignano = import_data('aquifer', 'petrignano')
bilancino = import_data('lake', 'bilancino')
arno = import_data('river', 'arno')
amiata = import_data('waterspring', 'amiata')
lupa = import_data('waterspring', 'lupa')
madonna = import_data('waterspring', 'madonna_di_canneto')
```

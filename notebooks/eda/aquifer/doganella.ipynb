```python
import pandas as pd
df = pd.read_csv('../../../data/raw/aquifer/doganella.csv')
print(df.shape)
df.head()
```

```python
df.columns.tolist()
```

```python
# plot features
import matplotlib.pyplot as plt
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
```

```python
fig, axs = plt.subplots(2, 1, figsize=(20,10))
axs[0].scatter(df['Date'], df['Rainfall_Monteporzio'], s=1.5)
axs[1].scatter(df['Date'], df['Rainfall_Velletri'], s=1.5)
plt.show()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(20,10))
axs[0].scatter(df['Date'], df['Temperature_Monteporzio'], s=1.5)
axs[1].scatter(df['Date'], df['Temperature_Velletri'], s=1.5)
plt.show()
```

```python
fig, axs = plt.subplots(8, 1, figsize=(20,30))
axs[0].scatter(df['Date'], df['Volume_Pozzo_1'], s=1.5)
axs[1].scatter(df['Date'], df['Volume_Pozzo_2'], s=1.5)
axs[2].scatter(df['Date'], df['Volume_Pozzo_3'], s=1.5)
axs[3].scatter(df['Date'], df['Volume_Pozzo_4'], s=1.5)
axs[4].scatter(df['Date'], df['Volume_Pozzo_5+6'], s=1.5)
axs[5].scatter(df['Date'], df['Volume_Pozzo_7'], s=1.5)
axs[6].scatter(df['Date'], df['Volume_Pozzo_8'], s=1.5)
axs[7].scatter(df['Date'], df['Volume_Pozzo_9'], s=1.5)
plt.show()
```

```python
# plot target
fig, axs = plt.subplots(5, 2, figsize=(20,30))
axs[0,0].scatter(df['Date'], df['Depth_to_Groundwater_Pozzo_1'].abs(), s=1.5)
axs[0,1].scatter(df['Date'], df['Depth_to_Groundwater_Pozzo_2'].abs(), s=1.5)
axs[1,0].scatter(df['Date'], df['Depth_to_Groundwater_Pozzo_3'].abs(), s=1.5)
axs[1,1].scatter(df['Date'], df['Depth_to_Groundwater_Pozzo_4'].abs(), s=1.5)
axs[2,0].scatter(df['Date'], df['Depth_to_Groundwater_Pozzo_5'].abs(), s=1.5)
axs[2,1].scatter(df['Date'], df['Depth_to_Groundwater_Pozzo_6'].abs(), s=1.5)
axs[3,0].scatter(df['Date'], df['Depth_to_Groundwater_Pozzo_7'].abs(), s=1.5)
axs[3,1].scatter(df['Date'], df['Depth_to_Groundwater_Pozzo_8'].abs(), s=1.5)
axs[4,0].scatter(df['Date'], df['Depth_to_Groundwater_Pozzo_9'].abs(), s=1.5)
plt.show()
```

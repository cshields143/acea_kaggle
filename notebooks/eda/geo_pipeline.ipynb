Acea provides geodata for different sites and facilities; this notebook exists to clean that geodata and restrict it to only the locations our datasets mention.

```python
import pandas as pd
df = pd.read_csv('../../data/geo/stazioni.csv', ';')
df
```

```python
# some features are more equal than others
keep = [
    'IDStazione',
    'Nome',
    'Comune',
    'Provincia',
    'Strumento',
    'Unita\' Misura',
    'LAT [o]',
    'LON [o]',
    'Quota'
]

df1 = df[keep]
```

```python
something='Chiascio'
df1[df1['Nome'].str.contains(something) | df1['Comune'].str.contains(something) | df1['Provincia'].str.contains(something)]
```

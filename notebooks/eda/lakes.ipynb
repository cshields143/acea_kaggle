```python
import pandas as pd
```

```python
df = pd.read_csv('../../data/raw/lake/bilancino.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(5, 1, figsize=(20,25), sharex=True, sharey=True)
axs[0].plot(df['Rainfall_S_Piero'])
axs[0].set_title('South Piero')
axs[1].plot(df['Rainfall_Mangona'])
axs[1].set_title('Mangona')
axs[2].plot(df['Rainfall_S_Agata'])
axs[2].set_title('South Agata')
axs[3].plot(df['Rainfall_Cavallina'])
axs[3].set_title('Cavallina')
axs[4].plot(df['Rainfall_Le_Croci'])
axs[4].set_title('Le Croci')
fig.suptitle('Rainfall', fontweight='bold')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(20,5))
ax.plot(df['Temperature_Le_Croci'])
fig.suptitle('Temperature at Le Croci', fontweight='bold')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(20,5))
ax.plot(df['Lake_Level'])
fig.suptitle('Lake Level', fontweight='bold')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(20,5))
ax.plot(df['Flow_Rate'])
fig.suptitle('Flow Rate', fontweight='bold')
plt.show()
```

```python
spiero = df[df['Rainfall_S_Piero'].isnull()]['Date']
```

```python
for _ in spiero:
    print(_)
```

```python
mangona = df[df['Rainfall_Mangona'].isnull()]['Date']
```

```python
(spiero == mangona).sum() == len(spiero)
```

```python
sagata = df[df['Rainfall_S_Agata'].isnull()]['Date']
cavallina = df[df['Rainfall_Cavallina'].isnull()]['Date']
lecroci = df[df['Rainfall_Le_Croci'].isnull()]['Date']

(spiero == sagata).sum() == len(spiero), (spiero == cavallina).sum() == len(spiero), (spiero == lecroci).sum() == len(spiero)
```

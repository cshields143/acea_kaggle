```python
import pandas as pd
df = pd.read_csv('../../../data/raw/waterspring/amiata.csv')
print(df.shape)
df.head()
```

```python
df.columns.tolist()
```

```python
df.isnull().sum()
```

```python
100 * df.isnull().sum() / df.shape[0]
```

```python
for c in df.columns:
    print(
        c,
        '...',
        df[~df[c].isnull()].iloc[0]['Date'].strftime('%D'),
        '--',
        df[~df[c].isnull()].iloc[-1]['Date'].strftime('%D')
    )
```

```python
from datetime import datetime
df = df[df['Date'] > datetime(2014, 12, 31)]
```

```python
df.isnull().sum()
```

```python
100 * df.isnull().sum() / df.shape[0]
```

```python
# rainfall data
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
import matplotlib.pyplot as plt
fig, axs = plt.subplots(5, 1, figsize=(20,25), sharex=True, sharey=True)
axs[0].scatter(df['Date'], df['Rainfall_Castel_del_Piano'], s=1.5)
axs[1].scatter(df['Date'], df['Rainfall_Abbadia_S_Salvatore'], s=1.5)
axs[2].scatter(df['Date'], df['Rainfall_S_Fiora'], s=1.5)
axs[3].scatter(df['Date'], df['Rainfall_Laghetto_Verde'], s=1.5)
axs[4].scatter(df['Date'], df['Rainfall_Vetta_Amiata'], s=1.5)
plt.show()
```

```python
# depth to groundwater, requires abs
fig, axs = plt.subplots(3, 1, figsize=(20,15))
axs[0].scatter(df['Date'], df['Depth_to_Groundwater_S_Fiora_8'].abs(), s=1.5)
axs[1].scatter(df['Date'], df['Depth_to_Groundwater_S_Fiora_11bis'].abs(), s=1.5)
axs[2].scatter(df['Date'], df['Depth_to_Groundwater_David_Lazzaretti'].abs(), s=1.5)
plt.show()
```

```python
# temperature
fig, axs = plt.subplots(3, 1, figsize=(20,15))
axs[0].scatter(df['Date'], df['Temperature_Abbadia_S_Salvatore'], s=1.5)
axs[1].scatter(df['Date'], df['Temperature_S_Fiora'], s=1.5)
axs[2].scatter(df['Date'], df['Temperature_Laghetto_Verde'], s=1.5)
plt.show()
```

```python
# flow rate, requires abs
fig, axs = plt.subplots(4, 1, figsize=(20,20))
axs[0].scatter(df['Date'], df['Flow_Rate_Bugnano'].abs(), s=1.5)
axs[1].scatter(df['Date'], df['Flow_Rate_Arbure'].abs(), s=1.5)
axs[2].scatter(df['Date'], df['Flow_Rate_Ermicciolo'].abs(), s=1.5)
axs[3].scatter(df['Date'], df['Flow_Rate_Galleria_Alta'].abs(), s=1.5)
plt.show()
```

```python
# uses zero for null:
#   Bugnano
#   Arbure ??
#   Ermicciolo
```

```python
fig, axs = plt.subplots(2, 1, figsize=(20,10))
isz = df['Flow_Rate_Bugnano'] == 0
axs[0].scatter(df[isz]['Date'], df[isz]['Flow_Rate_Bugnano'], s=1.5, c='C1')
axs[0].scatter(df[~isz]['Date'], df[~isz]['Flow_Rate_Bugnano'].abs(), s=1.5, c='C0')
isz = df['Flow_Rate_Ermicciolo'] == 0
axs[1].scatter(df[isz]['Date'], df[isz]['Flow_Rate_Ermicciolo'], s=1.5, c='C1')
axs[1].scatter(df[~isz]['Date'], df[~isz]['Flow_Rate_Ermicciolo'].abs(), s=1.5, c='C0')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(20,5))
isz = df['Flow_Rate_Arbure'] == 0
ax.scatter(df[isz]['Date'], df[isz]['Flow_Rate_Arbure'], s=1.5, c='C1')
ax.scatter(df[~isz]['Date'], df[~isz]['Flow_Rate_Arbure'].abs(), s=1.5, c='C0')
plt.show()
```

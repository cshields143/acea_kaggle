```python
import pandas as pd
df = pd.read_csv('../../../data/raw/river/arno.csv')
print(df.shape)
df.head()
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(7, 2, figsize=(28,28), sharex=True, sharey=True)
axs[0,0].plot(df['Rainfall_Le_Croci'])
axs[0,0].set_title('Le Croci')
axs[0,1].plot(df['Rainfall_Cavallina'])
axs[0,1].set_title('Cavallina')
axs[1,0].plot(df['Rainfall_S_Agata'])
axs[1,0].set_title('S Agata')
axs[1,1].plot(df['Rainfall_Mangona'])
axs[1,1].set_title('Mangona')
axs[2,0].plot(df['Rainfall_S_Piero'])
axs[2,0].set_title('S Piero')
axs[2,1].plot(df['Rainfall_Vernio'])
axs[2,1].set_title('Vernio')
axs[3,0].plot(df['Rainfall_Stia'])
axs[3,0].set_title('Stia')
axs[3,1].plot(df['Rainfall_Consuma'])
axs[3,1].set_title('Consuma')
axs[4,0].plot(df['Rainfall_Incisa'])
axs[4,0].set_title('Incisa')
axs[4,1].plot(df['Rainfall_Montevarchi'])
axs[4,1].set_title('Montevarchi')
axs[5,0].plot(df['Rainfall_S_Savino'])
axs[5,0].set_title('S Savino')
axs[5,1].plot(df['Rainfall_Laterina'])
axs[5,1].set_title('Laterina')
axs[6,0].plot(df['Rainfall_Bibbiena'])
axs[6,0].set_title('Bibbiena')
axs[6,1].plot(df['Rainfall_Camaldoli'])
axs[6,1].set_title('Camaldoli')
plt.xlim([1900, 8400])
plt.ylim([-10, 135])
plt.show()
```
